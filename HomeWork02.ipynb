8. А тепер згрупуйте дані по району (borough) і вкажіть, з якого пункту було здійснено найбільше поїздок.

```python
taxi.groupby('borough', as_index=False) \
    .agg({'pickups': 'sum'}) \
    .sort_values(['pickups'], ascending = False) \
    .head(1)
```

10. Продовжимо вивчення даних та подивимося на кількість поїздок у вихідні дні. Згрупуйте дані за двома ознаками: району міста і чи є день вихідним (колонки borough та hday). Порівняйте середню кількість поїздок, і назвіть райони, з яких у свята в середньому надходить більше замовлень, ніж у звичайні дні.

```python
# Групуємо за районом та чи є день вихідним
taxi.groupby(['borough','hday'], as_index=False) \
    .agg({'pickups': 'mean'}) \
    .sort_values(['pickups'], ascending = False)
```

11. Для кожного району порахуйте кількість поїздок по місяцях. Відсортуйте отримані значення за спаданням та збережіть результуючий датафрейм у pickups_by_mon_bor. Зверніть увагу, що підсумковий датасет повинен складатися з трьох колонок - pickup_month, borough, pickups.

```python
pickups_by_mon_bor = taxi.groupby(['pickup_month','borough'], as_index=False) \
                            .agg({'pickups': 'sum'}) \
                            .sort_values(['pickups'], ascending = False)
```
