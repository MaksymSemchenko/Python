7. –ù–∞—Å—Ç—É–ø–Ω–µ –∑–∞–≤–¥–∞–Ω–Ω—è: –∑'—è—Å—É–≤–∞—Ç–∏, –∑ —è–∫–æ–≥–æ —Ä–∞–π–æ–Ω—É –±—É–ª–æ –∑–¥—ñ–π—Å–Ω–µ–Ω–æ –Ω–∞–π–±—ñ–ª—å—à—É –∫—ñ–ª—å–∫—ñ—Å—Ç—å –ø–æ—ó–∑–¥–æ–∫ –∑–∞ –ø–µ—Ä—ñ–æ–¥.
–î–ª—è –ø–æ—á–∞—Ç–∫—É –ø–æ—Ä–∞—Ö—É–π—Ç–µ –∑–∞–≥–∞–ª—å–Ω—É –∫—ñ–ª—å–∫—ñ—Å—Ç—å –ø–æ—ó–∑–¥–æ–∫ (pickups), –±–µ–∑ –≥—Ä—É–ø–æ–≤–∞–Ω–Ω—è.

8. –ê —Ç–µ–ø–µ—Ä –∑–≥—Ä—É–ø—É–π—Ç–µ –¥–∞–Ω—ñ –ø–æ —Ä–∞–π–æ–Ω—É (borough) —ñ –≤–∫–∞–∂—ñ—Ç—å, –∑ —è–∫–æ–≥–æ –ø—É–Ω–∫—Ç—É –±—É–ª–æ –∑–¥—ñ–π—Å–Ω–µ–Ω–æ –Ω–∞–π–±—ñ–ª—å—à–µ –ø–æ—ó–∑–¥–æ–∫.

```python
taxi.groupby('borough', as_index=False) \
    .agg({'pickups': 'sum'}) \
    .sort_values(['pickups'], ascending = False) \
    .head(1)
```

9. ‚≠êÔ∏è–ó–∞–≤–¥–∞–Ω–Ω—è –∑—ñ –∑—ñ—Ä–æ—á–∫–æ—é!‚≠êÔ∏è
–©–µ –∫—ñ–ª—å–∫–∞ –∫–æ—Ä–∏—Å–Ω–∏—Ö –º–µ—Ç–æ–¥—ñ–≤ —É pandas! –ú–æ–∂–ª–∏–≤–æ, —É –ø–æ–ø–µ—Ä–µ–¥–Ω—å–æ–º—É —Å—Ç–µ–ø—É –≤–∏ –ø—Ä–æ—Å—Ç–æ –ø–æ–¥–∏–≤–∏–ª–∏—Å—è –Ω–∞ –¥–∞–Ω—ñ –∞–±–æ –≤—ñ–¥—Å–æ—Ä—Ç—É–≤–∞–ª–∏ –∑–Ω–∞—á–µ–Ω–Ω—è. –ü—Ä–∏—Å–∫–æ—Ä–∏—Ç–∏ —Ü–µ–π –ø—Ä–æ—Ü–µ—Å –Ω–∞–¥–∞–ª—ñ –¥–æ–ø–æ–º–æ–∂—É—Ç—å –º–µ—Ç–æ–¥–∏ idxmin() —Ç–∞ idxmax(), —è–∫—ñ –ø–æ–≤–µ—Ä—Ç–∞—é—Ç—å —ñ–Ω–¥–µ–∫—Å –º—ñ–Ω—ñ–º–∞–ª—å–Ω–æ–≥–æ —á–∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ –∑–Ω–∞—á–µ–Ω–Ω—è.
–ó–±–µ—Ä–µ–∂—ñ—Ç—å –Ω–∞–∑–≤—É —Ä–∞–π–æ–Ω—É –∑ –Ω–∞–π–º–µ–Ω—à–æ—é –∫—ñ–ª—å–∫—ñ—Å—Ç—é –ø–æ—ó–∑–¥–æ–∫ —É –∑–º—ñ–Ω–Ω—É min_pickups, –∑–∞—Å—Ç–æ—Å—É–≤–∞–≤—à–∏ –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω–∏–π –º–µ—Ç–æ–¥.

```python
taxi.groupby('borough') \
    .agg({'pickups': 'sum'}) \
    .idxmin()
```

10. –ü—Ä–æ–¥–æ–≤–∂–∏–º–æ –≤–∏–≤—á–µ–Ω–Ω—è –¥–∞–Ω–∏—Ö —Ç–∞ –ø–æ–¥–∏–≤–∏–º–æ—Å—è –Ω–∞ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –ø–æ—ó–∑–¥–æ–∫ —É –≤–∏—Ö—ñ–¥–Ω—ñ –¥–Ω—ñ. –ó–≥—Ä—É–ø—É–π—Ç–µ –¥–∞–Ω—ñ –∑–∞ –¥–≤–æ–º–∞ –æ–∑–Ω–∞–∫–∞–º–∏: —Ä–∞–π–æ–Ω—É –º—ñ—Å—Ç–∞ —ñ —á–∏ —î –¥–µ–Ω—å –≤–∏—Ö—ñ–¥–Ω–∏–º (–∫–æ–ª–æ–Ω–∫–∏ borough —Ç–∞ hday). –ü–æ—Ä—ñ–≤–Ω—è–π—Ç–µ —Å–µ—Ä–µ–¥–Ω—é –∫—ñ–ª—å–∫—ñ—Å—Ç—å –ø–æ—ó–∑–¥–æ–∫, —ñ –Ω–∞–∑–≤—ñ—Ç—å —Ä–∞–π–æ–Ω–∏, –∑ —è–∫–∏—Ö —É —Å–≤—è—Ç–∞ –≤ —Å–µ—Ä–µ–¥–Ω—å–æ–º—É –Ω–∞–¥—Ö–æ–¥–∏—Ç—å –±—ñ–ª—å—à–µ –∑–∞–º–æ–≤–ª–µ–Ω—å, –Ω—ñ–∂ —É –∑–≤–∏—á–∞–π–Ω—ñ –¥–Ω—ñ.

```python
# –ì—Ä—É–ø—É—î–º–æ –∑–∞ —Ä–∞–π–æ–Ω–æ–º —Ç–∞ —á–∏ —î –¥–µ–Ω—å –≤–∏—Ö—ñ–¥–Ω–∏–º
taxi.groupby(['borough','hday'], as_index=False) \
    .agg({'pickups': 'mean'}) \
    .sort_values(['pickups'], ascending = False)
```

11. –î–ª—è –∫–æ–∂–Ω–æ–≥–æ —Ä–∞–π–æ–Ω—É –ø–æ—Ä–∞—Ö—É–π—Ç–µ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –ø–æ—ó–∑–¥–æ–∫ –ø–æ –º—ñ—Å—è—Ü—è—Ö. –í—ñ–¥—Å–æ—Ä—Ç—É–π—Ç–µ –æ—Ç—Ä–∏–º–∞–Ω—ñ –∑–Ω–∞—á–µ–Ω–Ω—è –∑–∞ —Å–ø–∞–¥–∞–Ω–Ω—è–º —Ç–∞ –∑–±–µ—Ä–µ–∂—ñ—Ç—å —Ä–µ–∑—É–ª—å—Ç—É—é—á–∏–π –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —É pickups_by_mon_bor. –ó–≤–µ—Ä–Ω—ñ—Ç—å —É–≤–∞–≥—É, —â–æ –ø—ñ–¥—Å—É–º–∫–æ–≤–∏–π –¥–∞—Ç–∞—Å–µ—Ç –ø–æ–≤–∏–Ω–µ–Ω —Å–∫–ª–∞–¥–∞—Ç–∏—Å—è –∑ —Ç—Ä—å–æ—Ö –∫–æ–ª–æ–Ω–æ–∫ - pickup_month, borough, pickups.

```python
pickups_by_mon_bor = taxi.groupby(['pickup_month','borough'], as_index=False) \
                            .agg({'pickups': 'sum'}) \
                            .sort_values(['pickups'], ascending = False)
```

```python
pickups_by_mon_bor
```
